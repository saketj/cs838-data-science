```python
dict = {}
for line in open("../datasets/cuisines.txt"):
    cuisines = line.split(",")
    for cuisine in cuisines:
        dict[cuisine.strip()] = cuisines[0].strip();
target = open('../datasets/cuisines_dictionary.json', 'w')
json.dump(dict, target)
print("done")
```
