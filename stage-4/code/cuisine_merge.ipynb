## Prints unique cuisines split on space, then we merge the cuisines manually into categories, like pizza, pasta, italian are merged into one category called italian

```python
import re
import csv
import json

reader = csv.DictReader(open("../datasets/matched_A_B_positive.csv"))
uniq = set();
for x in reader:
    keys = ['ltable_cuisine','rtable_cuisine']
    for key in keys:   
        z = re.split("\W", x[key])
        for i in z:
            i = i.strip();
            if(i not in uniq):
                uniq.add(i)
                print(i)
```

## Converts our manually created cuisines.txt to a json map

```python
dict = {}
for line in open("../datasets/cuisines.txt"):
    cuisines = line.split(",")
    for cuisine in cuisines:
        dict[cuisine.strip()] = cuisines[0].strip();
target = open('../datasets/cuisines_dictionary.json', 'w')
json.dump(dict, target)
print("done")
```
